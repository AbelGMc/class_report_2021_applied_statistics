```python
import pyreadr
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct,Matern,WhiteKernel
```

```python
project_data = pyreadr.read_r('../Data/project_data.Rdata')
train_data = project_data["train_data"]
test_data = project_data["test_data"]
train_meta_data = project_data["train_meta_data"]
test_meta_data = project_data["test_meta_data"]
```

```python
train_data
```

```python
train_meta_data
```

```python
object_id = (train_meta_data.loc[train_meta_data['target']==88,"object_id"]).iloc[0]
index = (train_data['object_id']==object_id)
train_data_for_object = train_data.iloc[index.values,]
index = train_data_for_object['passband']==1
train_data_for_object = train_data_for_object.iloc[index.values,]
```

```python
kernel = 1.0 * Matern(length_scale=1.0, nu=1.5) + WhiteKernel()
X = np.array(train_data_for_object['mjd']).reshape(-1,1)
y =  train_data_for_object['flux']
gpr = GaussianProcessRegressor(kernel=kernel,random_state=0,n_restarts_optimizer=200).fit(X,y)
```

```python
x = np.atleast_2d(np.linspace(59800,60700, 1000)).T
y_pred, sigma = gpr.predict(x, return_std=True)
```

```python
plt.figure()
plt.plot(X, y, 'ro', label='observations')
plt.plot(x, y_pred, 'b-', label='Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                        (y_pred + 1.9600 * sigma)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
plt.xlabel('mjd')
plt.ylabel('flux for passband 1')
plt.legend(loc='upper left')
plt.title("Fitted flux for a source in class 88")
plt.savefig('../Images/proposal_img8.png')
```
